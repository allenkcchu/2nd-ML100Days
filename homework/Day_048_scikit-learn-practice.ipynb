```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import r2_score, accuracy_score

dir_data = './data/data-science-london-scikit-learn/'
trainData = os.path.join(dir_data, 'train.csv')
trainTarget = os.path.join(dir_data, 'trainLabels.csv')

data = np.array(pd.read_csv(trainData))
target = np.array(pd.read_csv(trainTarget))
[x_train,x_test,y_train,y_test] = train_test_split(data,target.ravel(),test_size=0.25,random_state=43)
```

```python
n_estimators = [100,200,300,400,500]
max_depth = [1,2,3,4,5]
learning_rate = [0.1,1,10,100,1000]
# subsample = 1-np.arange(0,1,0.1)
param_grid = dict(n_estimators = n_estimators, 
                  learning_rate = learning_rate, 
                  max_depth = max_depth)
grid_search = GridSearchCV(GradientBoostingClassifier(), param_grid, 
                                 scoring='neg_log_loss', n_jobs=-1, verbose=1, cv=5)
grid_result = grid_search.fit(x_train,y_train)
print('Best Accuracy: %.3f using %s' % (grid_result.best_score_, grid_result.best_params_))
clf = GradientBoostingClassifier(n_estimators = grid_result.best_params_['n_estimators'], 
                                 max_depth = grid_result.best_params_['max_depth'], 
                                 learning_rate = grid_result.best_params_['learning_rate'])
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
result = accuracy_score(y_test,y_pred)
print('Accuracy score of this regrssion problem is %.3f' % (result))
```

```python
test = os.path.join(dir_data, 'test.csv')
testData = pd.read_csv(test,header=None)
result = clf.predict(testData)
testData.head()
```

```python
df = pd.DataFrame({'Id':testData.index+1, 
                   'Solution':result})
df.to_csv('test1.csv',index=False)
df.head()
```
